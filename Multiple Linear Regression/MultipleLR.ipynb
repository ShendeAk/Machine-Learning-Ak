# Multiple Linear Regression

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

# Read the dataset

```python
import pandas as pd
df = pd.read_csv('50_Startups.csv')
df.head()
```

### PROFIT ~ RND, ADMIN, MKT
### PROFIT = B0 + B1xRND +B2xMKT+B3xADMIN

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

###  Seperating X and Y

```python
X = df[['RND', 'ADMIN', 'MKT']]
Y = df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

### Create a Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, Y)
```

### CHECK INTERCEPT AND COEFFICIENTS

```python
model.intercept_
```

```python
model.coef_
```

```python
X.columns
```

### PROFIT_pred = 50122.19 + 0.8257xRND - 0.268xADMIN + 0.0272xMKT

```python
coef = pd.Series(model.coef_.flatten())
coef.index = X.columns
coef
```

```python
coef.plot(kind='bar', title='Coeficent of Linear Reg')
```

1. IF RND INCRASE BY 1000, PROFIT INCRESES BY805.7
2. IF ADMIN INCREASE BY 1000, PROFIT DECREASE BY 26.
3. IF MKT INCREASE BY 1000, PROFIT INCREASE BY 27.2

### eVALUATE THE MODEL

```python
# R2 SCORE
model.score(X,Y)
```

### Predict the results for X 

```python
ypred = model.predict(X)
ypred[0:5]
```

```python
Y.head()
```

### MSE, RMSE, MAE, R2

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(Y, ypred)
mse
```

```python
rmse = mse**(1/2)
rmse
```

```python
mae = mean_absolute_error(Y, ypred)
mae
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(Y, ypred)
mape
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y, ypred)
r2
```
