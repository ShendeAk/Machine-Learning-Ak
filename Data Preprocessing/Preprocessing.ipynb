### To replace missing value function in pandas is df[column_name].fillna(value)

```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for col in df.columns:
        if col in cat:
            md = df[col].mode()[0] 
            df[col] = df[col].fillna(md)
        else:
            mn = df[col].mean()
            df[col] = df[col].fillna(mn)
    print('Missing values replaced')
```

### R2 score in test >0.8 hence above is good model

### Check mse, rmse, mae, r2

```python

def evaluate_model(model, x,y):
    # Predict the results for x
    ypred = model.predict(x)
    # import necessary package
    from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
    from sklearn.metrics import mean_absolute_percentage_error
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred) 
    r2 = r2_score(y, ypred)
    # Print the results
    print(f'Mean squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute E rror : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```
